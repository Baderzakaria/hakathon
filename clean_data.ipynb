```python
import pandas as pd 
```

```python
df = pd.read_csv('events.csv')
```

```python
df
```

```python
df['category_code'].fillna('no_category', inplace=True)
df['brand'].fillna('no_brand', inplace=True)
df
```

```python
total_rows = len(df)
no_category_count = df['category_code'].value_counts().get('no_category', 0)
percentage_no_category = (no_category_count / total_rows) * 100

# Calculate the percentage of 'no_brand' values
total_rows = len(df)
no_brand_count = df['brand'].value_counts().get('no_brand', 0)
percentage_no_brand = (no_brand_count / total_rows) * 100

print(f"Percentage of 'no_category' values: {percentage_no_category:.2f}%")
print(f"Percentage of 'no_brand' values: {percentage_no_brand:.2f}%")
```

```python
# Vérifier si la colonne 'event_time' contient des valeurs nulles
print(df['event_time'].isnull().sum())

# Remplacer les valeurs nulles par une chaîne vide pour éviter les erreurs lors de la séparation
df['event_time'].fillna('', inplace=True)

# Séparer la colonne 'event_time' en deux colonnes 'date' et 'heures'
df['date'] = df['event_time'].str.split(' ', expand=True)[0]
df['heures'] = df['event_time'].str.split(' ', expand=True)[1]

# Supprimer le fuseau horaire 'UTC' de la colonne 'heures' si nécessaire
df['heures'] = df['heures'].str.replace(' UTC', '')

# Optionnel: Supprimer la colonne 'event_time' originale
# df.drop(columns=['event_time'], inplace=True)
```
